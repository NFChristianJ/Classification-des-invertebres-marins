```python
import torch
import torchvision.transforms as transforms
from PIL import Image
import torch.nn as nn
from sklearn.metrics import classification_report, confusion_matrix
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import json
```

```python
load_path = r'C:\Users\Christian\Desktop\ClassificationInvertebresMarrins\model_vit.pth'

# Charger le modèle complet
model = torch.load(load_path)
model.eval()  
```

```python
# Définir les transformations pour les images
transform = transforms.Compose([
    transforms.Resize((224, 224)),  # Redimensionner à la taille utilisée pour l'entraînement
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# Fonction pour charger et transformer une image
def load_image(image_path):
    image = Image.open(image_path).convert('RGB')
    image = transform(image)
    image = image.unsqueeze(0)  # Ajouter une dimension pour le batch
    return image
```

```python
class_indice_path = r'C:\Users\Christian\Desktop\ClassificationInvertebresMarrins\class_indices.json'
with open(class_indice_path, 'r') as f:
    class_indices = json.load(f)
```

```python
class_names = {v: k for k, v in class_indices.items()}  # {indice: classe}
```

```python
# Chemin vers l'ensemble de test
directory = r'C:\Users\Christian\Desktop\ClassificationInvertebresMarrins\Dataset\test'
```

```python
# Initialiser les listes pour stocker les vraies étiquettes et les prédictions
y_true = []
y_pred = []

# Boucle à travers les images de l'ensemble de test
for class_index, class_name in class_names.items():
    class_dir = os.path.join(directory, class_name)
    if os.path.isdir(class_dir):
        for image_name in os.listdir(class_dir):
            image_path = os.path.join(class_dir, image_name)
            image = load_image(image_path)
            
            # Faire la prédiction
            with torch.no_grad():
                output = model(image)
                _, predicted_class = torch.max(output, 1)
            
            # Ajouter les étiquettes et les prédictions
            y_true.append(class_name)
            y_pred.append(class_names[predicted_class.item()])
```

```python
# Calculer le rapport de classification
print("Rapport de classification :")
print(classification_report(y_true, y_pred, target_names=list(class_names.values())))

# Calculer la matrice de confusion
cm = confusion_matrix(y_true, y_pred, labels=list(class_names.values()))
```
