```python
import os

#creation d'un nouveau dossier pour les images qui seront renommées
dossier_principal = r"C:\Users\Christian\Desktop\UE Projet\Dataset\train_small"
nouveau_dossier = os.path.join(dossier_principal, "..", "train_data")
os.makedirs(nouveau_dossier, exist_ok=True)
print(f"Le dossier {nouveau_dossier} a été créé avec succès.")
```

```python
import shutil
```

```python
# Parcourir les sous-dossiers de train_small
for root, dirs, files in os.walk(dossier_principal):
    for dir_name in dirs:
        sous_dossier_path = os.path.join(root, dir_name)
        
        # Initialiser le compteur d'image
        compteur = 1
        
        # Parcourir les fichiers dans chaque sous-dossier
        for file_name in os.listdir(sous_dossier_path):
            # Vérifier que c'est bien une image (par exemple, extension .jpg, .png, .jpeg)
            if file_name.endswith(('.jpg', '.png', '.jpeg')):
                # Chemin complet de l'image
                image_path = os.path.join(sous_dossier_path, file_name)
                
                # Nouveau nom de l'image
                nouveau_nom = f"{dir_name}_{compteur}.jpeg"  # Changer l'extension si nécessaire
                
                # Chemin complet de la nouvelle image
                nouveau_image_path = os.path.join(nouveau_dossier, nouveau_nom)
                
                # Copier l'image avec le nouveau nom
                shutil.copy2(image_path, nouveau_image_path)
                
                # Incrémenter le compteur
                compteur += 1
                
print("Toutes les images ont été copiées et renommées avec succès !")
```

```python
# Fonction pour compter les images dans un dossier donné (y compris les sous-dossiers)
def compter_images(dossier):
    compteur = 0
    for root, dirs, files in os.walk(dossier):
        for file in files:
            if file.lower().endswith(('.jpeg', '.jpg')):
                compteur += 1
    return compteur

# Fonction pour compter les images dans chaque sous-dossier
def compter_images_par_sous_dossier(dossier):
    resultats = {}
    for sous_dossier in os.listdir(dossier):
        chemin_sous_dossier = os.path.join(dossier, sous_dossier)
        if os.path.isdir(chemin_sous_dossier):
            compteur = 0
            for file in os.listdir(chemin_sous_dossier):
                if file.lower().endswith(('.jpeg', '.jpg')):
                    compteur += 1
            resultats[sous_dossier] = compteur
    return resultats

# Nombre total d'images dans train_small
total_images_train_small = compter_images(dossier_principal)

# Nombre total d'images dans train_data
total_images_train_data = compter_images(nouveau_dossier)

# Nombre total d'images dans chaque sous-dossier de train_small
images_par_sous_dossier = compter_images_par_sous_dossier(dossier_principal)

# Afficher les résultats
print(f"Nombre total d'images dans train_small : {total_images_train_small}")
print(f"Nombre total d'images dans train_data : {total_images_train_data}")
print("Nombre total d'images dans chaque sous-dossier de train_small :")
for sous_dossier, count in images_par_sous_dossier.items():
    print(f"{sous_dossier} : {count}")
```

```python
import pandas as pd
# Préparation des données pour le DataFrame et sauvegardons dans un fichier csv
data = [
    {"dossiers": "train_small", "nombre d'espèce": total_images_train_small},
    {"dossiers": "train_data", "nombre d'espèce": total_images_train_data}
]

# Ajouter les sous-dossiers et leurs nombres d'images
for sous_dossier, count in images_par_sous_dossier.items():
    data.append({"dossiers": sous_dossier, "nombre d'espèce": count})

# Création du DataFrame
df = pd.DataFrame(data)

# Sauvegarder dans un fichier Excel
#output_excel = r"C:\Users\Christian\Desktop\UE Projet\Dataset\image_counts.xlsx"
#df.to_excel(output_excel, index=False)

# Sauvegarder dans un fichier CSV
output_csv = r"C:\Users\Christian\Desktop\UE Projet\Dataset\DenombrementDesImages.csv"
df.to_csv(output_csv, index=False)

print(f"Les résultats ont été sauvegardés dans {output_csv}")
```
