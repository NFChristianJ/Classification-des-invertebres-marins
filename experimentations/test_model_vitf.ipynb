```python
import torch
from timm import create_model
import torch.nn as nn
import os
from torchvision import transforms
from PIL import Image
```

```python
load_path = r'C:\Users\Christian\Desktop\ClassificationInvertebresMarrins\model_vit.pth'

# Charger le modèle complet
model = torch.load(load_path)
model.eval()  
```

```python
directory = r'C:\Users\Christian\Desktop\ClassificationInvertebresMarrins\Dataset\test_small\test_small_1'

# Renommer les images avec des numéros séquentiels
for count, filename in enumerate(os.listdir(directory), start=1):
    src = os.path.join(directory, filename)
    dst = os.path.join(directory, f"{count}.jpg")
    os.rename(src, dst)
```

```python
# Définir les transformations pour les images
transform = transforms.Compose([
    transforms.Resize((224, 224)),  # Redimensionner à la taille utilisée pour l'entraînement
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# Fonction pour charger et transformer une image
def load_image(image_path):
    image = Image.open(image_path).convert('RGB')
    image = transform(image)
    image = image.unsqueeze(0)  # Ajouter une dimension pour le batch
    return image
```

```python
# Dictionnaire pour stocker les prédictions
results = {}

# Évaluer le modèle sur chaque image
with torch.no_grad():
    for filename in os.listdir(directory):
        if filename.endswith(".jpg"):
            image_path = os.path.join(directory, filename)
            image = load_image(image_path)
            output = model(image)
            _, predicted = torch.max(output, 1)
            class_name = predicted.item()  # Obtenir l'indice de la classe prédite
            results[filename] = class_name
```

```python
# Afficher les résultats
for filename, class_name in results.items():
    print(f"L'image {filename} correspond à la classe {class_name}")
```
